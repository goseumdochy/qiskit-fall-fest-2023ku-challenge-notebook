## Part 4: Inner & Outer Product

ì›Œí¬ìˆì—ì„œ inner productì™€ outer productì— ëŒ€í•´ì„œ ë‹¤ë£¨ì—ˆìŠµë‹ˆë‹¤. íŒŒì´ì¬ì—ì„œëŠ” inner productì™€ outer productë¥¼ numpyì˜ `.dot()` ë§¤ì„œë“œì™€ `.outer()`ë§¤ì„œë“œë¥¼ ì´ìš©í•˜ì—¬ í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.
